```python
# Load the pickle file
filename = 'results_maxDice6-maxRerolls5.pkl'
with open(filename, 'rb') as f:
    data = pickle.load(f)
data
```

```python
def change_combination_tuple_to_string(tuple_combination):
    my_str = str(tuple_combination)
    translation_table = {
        '(': '',
        ')': '',
        ', ': '-',
        ',': '',
        '\'': '',
        'white': 'w',
        'yellow': 'y',
        'red': 'r',
        'black': 'b'
    }
    for key, value in translation_table.items():
        my_str = my_str.replace(key, value)
    return my_str
```

```python
# Assemble a huge dataframe
# Initialize an empty DataFrame to collect results
results_df = pd.DataFrame()

# Loop through each entry in the data array
for entry in data:
    result_df = entry['result_df']
    temp_df = result_df[['ccdf']].copy()
    temp_df = temp_df.T
    
    temp_df.reset_index(drop=True, inplace=True)
    temp_df['combination'] = change_combination_tuple_to_string(entry['combination'])
    temp_df['rerolls'] = entry['rerolls']
    temp_df['miss_rate'] = entry['miss_rate']
    
    # Append to the results DataFrame
    results_df = pd.concat([results_df, temp_df], ignore_index=True)

imp_columns = ['combination', 'rerolls', 'miss_rate']
columns = [col for col in results_df.columns if col not in imp_columns]
columns = imp_columns + columns
results_df = results_df[columns]
results_df.fillna(0, inplace=True)
results_df

  
```

```python
# Save the results to a CSV file
results_df.to_csv('summary_results_maxDice6-maxRerolls5.csv', index=False)
```
